# Matplotlib Example Workflow

```python
# First import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]
# Set up plot
fig,ax = plt.subplots(figsize=(10,10))
# Plot the data
ax.plot(x,y)
# Customize plot
ax.set(title="Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")
# Save and show the fig
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot from a dictionary

test_dict = {"one": 5, "two": 10, "three": 30}

fig, ax = plt.subplots()
ax.bar(test_dict.keys(), test_dict.values())
ax.set(title="Test Bar Chart");
```

```python
# horizontal bar chart
fig, ax = plt.subplots()
ax.barh(list(test_dict.keys()),list(test_dict.values()))
ax.set(title="Test Bar Chart");
```

```python
# histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
ax.set(title="Test Histogram");
```

# Two options for subplots

```python
# Subplot option 1
fig, ((ax0,ax1), (ax2,ax3)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax0.plot(x,x)
ax1.scatter(x,x/10)
ax2.bar(test_dict.keys(),test_dict.values())
ax3.hist(x);
```

```python
#Subplot option 2

fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax[0,0].plot(x,x);
ax[0,1].scatter(x,x/10);
ax[1,0].bar(test_dict.keys(),test_dict.values())
ax[1,1].hist(x);
```

## Which one should you use? pyplot vs matplotlib OO method?

* When plotting something quickly - okay to use pyplot method.

* When plotting something more advanced - use the OO method


```python
over_50 = heart_disease[heart_disease["age"] > 50] # boolean indexing
over_50
```

```python
## pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
## OO method mixed with pyplot method

fig,ax = plt.subplots(figsize=(20,10))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);

ax.set_xlim(50,80);
```

```python
## OO method from scratch

fig,ax = plt.subplots(figsize=(20,10))

scatter_plot = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

ax.set(title = "Heart Disease and Cholesterol Levels", xlabel='Age', ylabel='Cholesterol');

ax.legend(*scatter_plot.legend_elements(), title="Target");

ax.axhline(y = over_50['chol'].mean(), linestyle = '--');
```

```python
#Create our figure and axis
plt.style.use('seaborn')
fig, ax = plt.subplots(nrows = 2, figsize = (10,10), sharex = True);

#Add data
scatter0 = ax[0].scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);

#Customize axes 0
ax[0].set(title="Chol and Age", xlabel = 'Age', ylabel = 'Chol');

#Add legend to axis 0 
ax[0].legend(*scatter0.legend_elements(), title = 'Target');

#Add meanline
ax[0].axhline(y=over_50['chol'].mean(),linestyle='--')


#Add data
scatter1 = ax[1].scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target']);

#Customize axes 0
ax[1].set(title="Thalach and Age", xlabel = 'Age', ylabel = 'thalach');

#Add legend to axis 0 
ax[1].legend(*scatter0.legend_elements(), title = 'Target');

#Add meanline
ax[1].axhline(y=over_50['thalach'].mean(),linestyle='--');

#Add a title to the figure

fig.suptitle('Heart Disease Data Exploration', size = 16)



```

# Customizing your plots

```python
plt.style.available
```

```python
#Create our figure and axis
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(nrows = 2, figsize = (10,10), sharex = True);
#Add data
scatter0 = ax[0].scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='winter');
#Customize axes 0
ax[0].set(title="Chol and Age", xlabel = 'Age', ylabel = 'Chol');
#Add legend to axis 0 
ax[0].legend(*scatter0.legend_elements(), title = 'Target');
#Add meanline
ax[0].axhline(y=over_50['chol'].mean(),linestyle='--')


#Add data
scatter1 = ax[1].scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap='winter');
#Customize axes 0
ax[1].set(title="Thalach and Age", xlabel = 'Age', ylabel = 'thalach');
#Add legend to axis 0 
ax[1].legend(*scatter0.legend_elements(), title = 'Target');
#Add meanline
ax[1].axhline(y=over_50['thalach'].mean(),linestyle='--');
#Add a title to the figure
fig.suptitle('Heart Disease Data Exploration', size = 16)

#Format axis
ax[0].set_xlim(50,80);
ax[1].set_ylim(70,200);
```
